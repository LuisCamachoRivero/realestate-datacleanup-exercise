```python
# TODO
south_ds.loc[ : , 'pps'] = south_ds.loc[ : , 'price'] / south_ds.loc[ : , 'surface']

getafe_filter = south_ds['level5'].str.contains('Getafe', na=False)
getafe_filtered = south_ds[getafe_filter]
getafe_mean_m2 = getafe_filtered['pps'].mean()


alcorcon_filter = south_ds['level5'].str.contains('Alcorcón', na=False)
alcorcon_filtered = south_ds[alcorcon_filter]
alcorcon_mean_m2 = alcorcon_filtered['pps'].mean()

new_south_ds = pd.concat([getafe_filtered, alcorcon_filtered], axis=0)

print (new_south_ds[['level5','price','surface','pps']].head())
print (f"\nMedia Pps Alcorcón:{round(alcorcon_mean_m2,2)} \nMedia Pps Getafe: {round(getafe_mean_m2,2)}")
print (f"Diferencia: {round(alcorcon_mean_m2 - getafe_mean_m2,2)}")

mean_pps = new_south_ds.groupby('level5')['pps'].mean().reset_index()

plt.bar(mean_pps['level5'], mean_pps['pps'])
plt.xlabel('Ubicación')
plt.ylabel('Promedio del Pps')
plt.title('Promedio de Pps de Casas por Ubicación')
plt.show()

plt.figure(figsize = (10, 5))
plt.hist(x=alcorcon_filtered['pps'], bins = 100, alpha = 0.7)
plt.title("Histograma Pps de Alcorcón")
plt.show()

plt.figure(figsize = (10, 5))
plt.hist(x=getafe_filtered['pps'], bins = 100, alpha = 0.7)
plt.title("Histograma Pps de Getafe")
plt.show()
```
